## Combining all data

```python
#import libraries

import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_ahmedabad = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_ahmedabad_hotels.csv')
data_bangalore = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_bangalore_hotels.csv')
data_chennai = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_chennai_hotels.csv')
data_delhi = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_delhi_hotels.csv')
data_hyderabad = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_hyderabad_hotels.csv')
data_kolkata = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_kolkata_hotels.csv')
data_mumbai = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_mumbai_hotels.csv')
data_pune = pd.read_csv(r'/Users/prashiklamsoge/Desktop/Project CSV File/yatra_pune_hotels.csv')
```

```python
data_ahmedabad['City'] = 'Ahmedabad'
data_bangalore['City'] = 'Bangalore'
data_chennai['City'] = 'Chennai'
data_delhi['City'] = 'Delhi'
data_hyderabad['City'] = 'Hyderabad'
data_kolkata['City'] = 'Kolkata'
data_mumbai['City'] = 'Mumbai'
data_pune['City'] = 'Pune'
```

```python
data['Price'] = data['Price'].astype(str).str.replace('[^0-9.]','', regex=True).astype(float)
data['Discount Percentage'] = data['Discount Percentage'].astype(str).str.replace('[^0-9.]','', regex=True).astype(float)
```

```python
# Featured Engineering

# Price Category (Budget / Mid / Luxury) 

def price_category(price):
    if price <= 2000:
        return "Budget"
    elif price <= 5000:
        return "Mid-range"
    else:
        return "Luxury"

data["Price_Category"] = data["Price"].apply(price_category)
```

```python
data
```

```python
data["Price_Category"].value_counts()
```

```python
# Price Band
data["Price_Band"] = pd.cut(
    data["Price"],
    bins=[0, 2000, 5000, 10000, np.inf],
    labels=["0-2K", "2K-5K", "5K-10K", "10K+"]
)
```

```python
# Price Distribution – Histogram

plt.figure(figsize=(8,5))
plt.hist(data['Price'], bins=30)
plt.xlabel("Price per Night (₹)")
plt.ylabel("Number of Hotels")
plt.title("Distribution of Hotel Prices")
plt.tight_layout()
plt.show()
```

```python
# Rating Distribution --
plt.figure(figsize=(8,5))
plt.hist(data['Rating'], bins=30)
plt.xlabel("Price per Night (₹)")
plt.ylabel("Number of Hotels")
plt.title("Distribution of Hotel Prices w.r.t. Ratings")
plt.tight_layout()
plt.show()
```

```python
# Hotels per City

plt.figure(figsize=(8,5))
data['City'].value_counts().plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Number of Hotels")
plt.title("Number of Hotels per City")
plt.tight_layout()
plt.show()
```

```python
# Average price per city

plt.figure(figsize=(8,5))
data.groupby('City')['Price'].mean().plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Average Price per Night (₹)")
plt.title("Average Hotel Price per City")
plt.tight_layout()
plt.show()
```

```python
# Discount vs Price

plt.figure()
plt.scatter(data['Discount Percentage'], data['Price'])
plt.xlabel("Discount Percentage")
plt.ylabel("Price (₹)")
plt.title("Discount vs Price")
plt.show()
```

```python
# Location wise Insights

# Top 10 Expensive Locations

data.groupby('Location')['Price'].mean().sort_values(ascending=False).head(10)
```

```python
# Top 10 Rated Loactions

data.groupby('Location')['Rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Correlation HeatMap

plt.figure()
sns.heatmap(data[['Price','Rating','Discount Percentage']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Price Outliers

plt.figure()
plt.boxplot(data['Price'])
plt.title("Price Outliers")
plt.ylabel("Price (₹)")
plt.show()
```

```python
cities = data['City'].unique()

for city in cities:
    city_data = data[data['City'] == city]
    
    plt.figure()
    plt.hist(city_data['Price'], bins=20)
    plt.title(f"Price Distribution - {city}")
    plt.xlabel("Price")
    plt.ylabel("Hotels")
    plt.tight_layout()
    plt.show()
```
