## Pune City

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome()
  
```

```python
url = "https://hotel.yatra.com/nextui/hotel-search/dom/search?checkoutDate=04%2F02%2F2026&checkinDate=03%2F02%2F2026&source=BOOKING_ENGINE&pg=1&tenant=B2C&isPersnldSrp=1&city.name=Pune&city.code=Pune&state.name=Maharashtra&state.code=Maharashtra&country.name=IND&country.code=IND&roomRequests%5B0%5D.id=1&roomRequests%5B0%5D.noOfAdults=2&roomRequests%5B0%5D.noOfChildren=0"
driver.get(url)
time.sleep(8) 
```

```python
for _ in range(50):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(4)
```

```python
soup = BeautifulSoup(driver.page_source, "lxml")

hotels = soup.find_all("div", class_="HotelListCard_hotelCardWrapper__krJC3 false")
len(hotels)
```

```python
data = []

for hotel in hotels:
    # Hotel Name
    try:
        name = hotel.find("h2").get_text(strip=True)
    except:
        name = None

    # Location
    try:
        location = hotel.find("span", class_="HotelListCard_addressText__XNcuF").get_text(strip=True)
    except:
        location = None

    # Price
    try:
        price = hotel.find("h2", class_="Typography_typography__ixark undefined Typography_h2__ie4tH HotelListCard_totalPrice__uISQ_").get_text(strip=True)
    except:
        price = None

    # Rating
    try:
        rating = hotel.find("span", class_="undefined ReviewRating_ratingValue__1fR09 ReviewRating_highRatingBg__mDpP1").get_text(strip=True)
    except:
        rating = None


    # discount percentage
    try:
        discount_percentage = hotel.find("span", class_="HotelListCard_discount__4CMC8").get_text(strip=True)
    except:
        rating = None

    data.append([name, location, price, rating, discount_percentage])
```

```python
df = pd.DataFrame(
    data,
    columns=["Hotel Name", "Location", "Price", "Rating", "Discount Percentage"]
)

df.head()
```

```python
df.to_csv("yatra_pune_hotels.csv", index=False)
```
